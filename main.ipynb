```python
class myCallback(Callback):
    def on_epoch_end(self, epoch, logs = {}):
        if logs.get('accuracy') > 0.85:
            self.model.stop_training = True
callback = myCallback()
# earlyStopping = EarlyStopping(patience=100, monitor = 'accuracy', restore_best_weights = True)
```

```python
model_nn = tf.keras.Sequential([tf.keras.layers.Dense(256, activation = tf.nn.relu, kernel_regularizer = regularizers.l2(0.01)),
                                tf.keras.layers.Dropout(0.1),
                                tf.keras.layers.Dense(256, activation = tf.nn.tanh, kernel_regularizer = regularizers.l2(0.01)),
                                tf.keras.layers.Dropout(0.1),
                                tf.keras.layers.Dense(256, activation = tf.nn.tanh, kernel_regularizer = regularizers.l2(0.01)),
                                tf.keras.layers.Dense(1, activation = tf.nn.sigmoid)])
model_nn.compile(optimizer = tf.optimizers.RMSprop(learning_rate=0.01),
              loss = tf.losses.binary_crossentropy,
              metrics = ['accuracy'])
model_nn.fit(trn, trl, epochs = 500, callbacks = [callback])
model_nn.evaluate(ten, tel)
```

```python
def rounding_nn_BinaryPred(pred_nn = [[]]):
    for i in range(len(np.squeeze(pred_nn))):
        if pred_nn[i]<0.5:
            pred_nn[i] = np.floor(pred_nn[i])
        else:
            pred_nn[i] = np.ceil(pred_nn[i])
```
